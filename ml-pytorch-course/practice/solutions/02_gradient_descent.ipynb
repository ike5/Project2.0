## Drill 1 — Hand-rolled GD in 1-D

Minimize a hidden 1-D quadratic with your own gradient-descent loop. Tune the learning rate / steps so `w_final` reaches the minimizer.

```python
task = pu.quadratic_task(dim=1)
task.describe()
```

```python
# ✅ reference solution
w = task.start.clone().requires_grad_(True)
for _ in range(800):
    task.loss(w).backward()
    with torch.no_grad():
        w -= 0.08 * w.grad
    w.grad.zero_()
w_final = w.detach()
```

```python
# check (run after your code)
task.grade(w_final)
```

## Drill 2 — GD in 2-D

Same idea, two parameters. The minimizer is a hidden 2-vector; descend to it.

```python
task = pu.quadratic_task(dim=2)
task.describe()
```

```python
# ✅ reference solution
w = task.start.clone().requires_grad_(True)
for _ in range(1500):
    task.loss(w).backward()
    with torch.no_grad():
        w -= 0.05 * w.grad
    w.grad.zero_()
w_final = w.detach()
```

```python
# check (run after your code)
task.grade(w_final)
```

## Drill 3 — Let torch.optim do the stepping

Re-solve the quadratic, but use a **real optimizer** (`torch.optim.SGD` or `Adam`) instead of hand-updating. Same `zero_grad → backward → step` ritual.

```python
task = pu.quadratic_task(dim=2)
task.describe()
```

```python
# ✅ reference solution
w = task.start.clone().requires_grad_(True)
opt = torch.optim.Adam([w], lr=0.1)
for _ in range(1000):
    opt.zero_grad()
    task.loss(w).backward()
    opt.step()
w_final = w.detach()
```

```python
# check (run after your code)
task.grade(w_final)
```

## Drill 4 — Pick a learning rate that converges

A quadratic's curvature is hidden. If your lr is too big you'll diverge; too small and 200 steps won't reach the minimum. Find an lr that lands `w_final` on the minimizer in **just 200 steps**.

```python
task = pu.quadratic_task(dim=1)
task.describe()
```

```python
# ✅ reference solution
lr = 0.2
w = task.start.clone().requires_grad_(True)
for _ in range(200):
    task.loss(w).backward()
    with torch.no_grad():
        w -= lr * w.grad
    w.grad.zero_()
w_final = w.detach()
```

```python
# check (run after your code)
task.grade(w_final)
```
