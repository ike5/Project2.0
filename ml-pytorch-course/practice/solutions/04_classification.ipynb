## Drill 1 — Classify random blobs

The task makes **2–4 Gaussian clusters** at fresh positions each draw. Build an MLP whose output is logits `(N, n_classes)` and train it to ≥ 90% on the hidden test set. Remember: final layer width = number of classes; use `CrossEntropyLoss`.

```python
task = pu.classification_task()
task.describe()
```

```python
# ✅ reference solution
Xtr, ytr = task.train_data()
n_out = int(ytr.max().item()) + 1
mx, sx = Xtr.mean(0, keepdim=True), Xtr.std(0, keepdim=True)
net = nn.Sequential(nn.Linear(2,32), nn.ReLU(), nn.Linear(32,32), nn.ReLU(), nn.Linear(32,n_out))
opt = torch.optim.Adam(net.parameters(), lr=0.02)
lf = nn.CrossEntropyLoss()
Xs = (Xtr - mx) / sx
for _ in range(300):
    opt.zero_grad()
    lf(net(Xs), ytr).backward()
    opt.step()
class Wrapped(nn.Module):
    def forward(self, X):
        return net((X - mx) / sx)
model = Wrapped()
```

```python
# check (run after your code)
task.grade(model)
```
