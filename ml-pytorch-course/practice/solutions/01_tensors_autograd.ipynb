## Drill 2 — Derivative with autograd

The task gives polynomial coefficients `[c0,c1,c2,c3]` (so `f(x)=c0+c1·x+c2·x²+c3·x³`) and a point `x0`. Use **autograd** to compute `df/dx` at `x0` into `student_grad`.

```python
coeffs, x0 = pu.random_poly()
print('coeffs', coeffs, ' x0', x0)
```

```python
# ✅ reference solution
x = torch.tensor(x0, requires_grad=True)
f = coeffs[0] + coeffs[1]*x + coeffs[2]*x**2 + coeffs[3]*x**3
f.backward()
student_grad = x.grad.item()
```

```python
# check (run after your code)
pu.grade_derivative(coeffs, x0, student_grad)
```

## Drill 3 — Standardize the columns

Standardize matrix `X` so every **column** has mean ≈ 0 and std ≈ 1, into `Z` (use `dim=0, keepdim=True` and broadcasting).

```python
X = torch.randn(200, 4) * torch.tensor([5., 0.1, 20., 1.]) + torch.tensor([3., -1., 50., 0.])
print('raw col means', [round(v,2) for v in X.mean(0).tolist()])
```

```python
# ✅ reference solution
Z = (X - X.mean(0, keepdim=True)) / X.std(0, keepdim=True)
```

```python
# check (run after your code)
g = pu.Grader('standardize')
g.at_most(Z.mean(0).abs().max().item(), 1e-4, 'max |col mean|')
g.almost(Z.std(0).mean().item(), 1.0, 1e-2, 'avg col std'); g.summary()
```

## Drill 4 — Gradient of a vector loss

For `loss = (w**2).sum()` the gradient is `2*w`. Fill `w.grad` via autograd into `student_grad` and match it.

```python
w = torch.randn(5, requires_grad=True)
print('w', [round(v,3) for v in w.tolist()])
```

```python
# ✅ reference solution
loss = (w**2).sum()
loss.backward()
student_grad = w.grad
```

```python
# check (run after your code)
g = pu.Grader('vector grad')
g.at_most((student_grad - 2*w).abs().max().item(), 1e-5, 'max error vs 2*w'); g.summary()
```

## Drill 5 — Minimize a random quadratic (by hand)

Minimize a quadratic with a **hidden** minimizer using gradient descent you write yourself (autograd for the gradient, `w -= lr*w.grad` under `no_grad`). Reach the min into `w_final`.

```python
task = pu.quadratic_task(dim=1)
task.describe()
```

```python
# ✅ reference solution
w = task.start.clone().requires_grad_(True)
for _ in range(500):
    loss = task.loss(w)
    loss.backward()
    with torch.no_grad():
        w -= 0.1 * w.grad
    w.grad.zero_()
w_final = w.detach()
```

```python
# check (run after your code)
task.grade(w_final)
```
